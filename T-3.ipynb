Task 3: Exploratory Data Analysis (EDA)

1.Plot distribution of numerical features using histograms.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("tv_shows.csv")
df.head()
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_cols:
    plt.figure()
    plt.hist(df[col].dropna(), bins=20)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

2.Analyze categorical features using count plots.

```python
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    plt.figure()
    df[col].value_counts().head(10).plot(kind='bar')
    plt.title(f"Top Categories in {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

3.Use box plots to identify outliers.

```python
for col in numerical_cols:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f"Box Plot of {col}")
    plt.show()
```

```python
4.Plot correlation heatmap to understand feature relationships.
```

```python
if 'duration' in df.columns:
    df['duration_num'] = df['duration'].str.extract(r'(\d+)').astype(float)

num_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure()
sns.heatmap(num_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
